```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a= np.arange(-1,1,0.1)
b= np.arange(-1,1,0.1)
a,b= np.meshgrid(a,b)
```

# SurfacePlot

```python
fig=plt.figure()
axes= fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```

# CONTOUR PLOT

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour3D(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```
